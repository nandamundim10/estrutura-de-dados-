# Laboratório: Explorando o `collections.deque`
**Objetivo:** Compreender na prática como usar a estrutura duplamente terminada `deque` do Python nativo para emular Filas (FIFO), Pilhas (LIFO) e Buffers contínuos de máxima eficiência de memória ($O(1)$).

```python
from collections import deque
import time
```

## 1. Exemplo como Fila (FIFO) - Pipeline de Dados
A Fila possui a política do "Primeiro a Entrar, Primeiro a Sair". Usamos o `append()` para adicionar no final e o `popleft()` para retirar do início de forma eficiente.

```python
print("--- COMPORTAMENTO DE FILA (FIFO) ---")

fila_pipeline = deque()

fila_pipeline.append("lote_treino_01.csv")
fila_pipeline.append("lote_treino_02.csv")
fila_pipeline.append("lote_treino_03.csv")

print(f"Fila de lotes aguardando: {fila_pipeline}")

lote_atual = fila_pipeline.popleft()
print(f"Processando modelo com: {lote_atual}")
print(f"Fila restante: {fila_pipeline}\n")
```

## 2. Exemplo como Pilha (LIFO) - Histórico/Backtracking

```python
print("--- COMPORTAMENTO DE PILHA (LIFO) ---")

historico_operacoes = deque()

historico_operacoes.append("Drop_NAs")
historico_operacoes.append("StandardScaler")
historico_operacoes.append("PCA")

print(f"Passos aplicados (Pilha): {historico_operacoes}")

ultimo_passo = historico_operacoes.pop()
print(f"Desfazendo a última operação: {ultimo_passo}")
print(f"Estado atual da Pilha: {historico_operacoes}\n")
```

## 3. Bônus: Limite Máximo de Memória (Buffer Circular)

```python
print("--- DEQUE COM TAMANHO MÁXIMO (Buffer Circular) ---")

logs_recentes = deque(maxlen=3)

logs_recentes.append("Log 1: Início")
logs_recentes.append("Log 2: Conexão DB")
logs_recentes.append("Log 3: Query executada")
print(f"Buffer cheio: {logs_recentes}")

logs_recentes.append("Log 4: Erro de timeout")
print(f"Buffer atualizado (Log 1 foi descartado): {logs_recentes}")
```

## 4. O Poder da Rotação (Listas Circulares)

```python
print("--- ROTAÇÃO DE ALTA PERFORMANCE ---")

jogadores = deque(["Alice", "Bruno", "Carlos", "Diana"])
print(f"Ordem inicial: {jogadores}")

jogadores.rotate(-1)
print(f"Após passar a vez do primeiro (rotate -1): {jogadores}")

jogadores.rotate(2)
print(f"Após girar 2 posições pra direita (rotate +2): {jogadores}")
```

## 5. Questão Discursiva
No bloco 1, nós criamos um `deque` chamado `fila_pipeline` para armazenar `lote_treino.csv`. Caso estivéssemos utilizando uma lista comum do Python (`list`) e utilizássemos o comando `.pop(0)` para remover o primeiro elemento, qual seria a complexidade de tempo (Big-O) dessa operação e qual seria o impacto físico na memória RAM do seu computador? Justifique detalhadamente por que o `deque.popleft()` resolve esse problema.

**Sua Resposta:**

Usar `list.pop(0)` tem complexidade **O(n)**.

**Por quê?** A `list` do Python é um array contíguo na memória RAM — todos os elementos ficam em endereços sequenciais adjacentes. Ao remover o elemento do índice `0`, cria-se um "buraco" no início desse bloco contíguo. Para preservar a integridade da estrutura, o Python precisa **deslocar fisicamente todos os `n-1` elementos restantes uma posição para a esquerda** na memória. Com 3 lotes isso é imperceptível, mas em produção com milhões de registros, esse deslocamento em cascata bloqueia o processador e degrada o sistema.

**Como o `deque.popleft()` resolve?** O `collections.deque` é implementado internamente como uma **lista duplamente encadeada** (doubly linked list). Cada nó guarda apenas um ponteiro para o próximo e um para o anterior — não há bloco contíguo de memória. Remover o nó da cabeça da fila é simplesmente atualizar dois ponteiros (`head = head.next`), sem mover dado algum. Isso garante complexidade **O(1)** constante, independente do tamanho da fila.

## 6. Trabalho Para Casa (TPC): A Fila do Call Center

```python
from collections import deque

fila_atendimento = deque()

print("=== Sistema de Fila do Call Center ===")
print("Digite o nome do cliente para enfileirar, ou 'sair' para encerrar.\n")

while True:
    nome = input("Nome do cliente: ").strip()

    if nome.lower() == 'sair':
        print("Sistema encerrado.")
        break

    if nome:
        fila_atendimento.append(nome)
        print(f"✅ '{nome}' adicionado à fila. Fila atual: {list(fila_atendimento)}")

    atendente_livre = input("Um atendente ficou livre? (S/N): ").strip().upper()

    if atendente_livre == 'S':
        if fila_atendimento:
            cliente_atendido = fila_atendimento.popleft()  # FIFO: remove o mais antigo
            print(f"📞 Atendendo: {cliente_atendido}")
            print(f"Fila restante: {list(fila_atendimento)}\n")
        else:
            print("ℹ️ Nenhum cliente aguardando na fila.\n")
```

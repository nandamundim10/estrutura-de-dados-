### Situação 1: O Botão Desfazer
Imagine as duas situações abaixo:
1. **A Fila do Banco:** O primeiro cliente que entra na fila é o primeiro a ser atendido pelo caixa.
2. **A Pilha de Pratos no Refeitório:** O último prato limpo colocado no topo da pilha é o primeiro a ser pego pelo próximo aluno.

**Pergunta:** Quando você digita um texto no Word, comete um erro e aperta `Ctrl+Z` (Desfazer), o computador está usando a lógica da Fila do Banco ou da Pilha de Pratos? Responda na célula abaixo justificando sua escolha.

**Sua Resposta:**

O `Ctrl+Z` usa a lógica da **Pilha de Pratos (LIFO — Last-In, First-Out)**.

Cada ação que o usuário realiza (digitar uma letra, apagar uma palavra, aplicar negrito) é empilhada no topo do histórico. Quando o `Ctrl+Z` é pressionado, o computador acessa o **topo da pilha** e desfaz a **última** ação registrada — exatamente como pegar o prato que foi colocado por último. A ação mais antiga fica no fundo e só é acessada depois que todas as ações mais recentes forem desfeitas.

### Situação 2: Como o Python faz isso?
A classe nativa `list` do Python já possui comandos embutidos para adicionar e remover dados. Vamos testá-la.
1. Crie uma lista vazia chamada `historico`.
2. Use o comando `.append()` para adicionar as strings: `"Passo 1"`, `"Passo 2"` e `"Passo 3"`.
3. Use o comando `.pop()` (sem nenhum argumento) e guarde o resultado numa variável chamada `removido`. Imprima a lista e o item removido.

```python
# Escreva seu código aqui
historico = []
historico.append("Passo 1")
historico.append("Passo 2")
historico.append("Passo 3")

removido = historico.pop()

print("Item desfeito:", removido)
print("Restante na memória:", historico)
```

### 🛑 O Problema do Atendimento
Agora, imagine que em vez de um "Desfazer", nossa lista represente uma fila de 10 milhões de requisições de clientes esperando atendimento em um servidor.

Para atender o **primeiro** cliente da fila (o índice `0`), um programador inexperiente usou o comando: `historico.pop(0)`.

**Pergunta Desafio:** Resgatando o que vimos na Aula 04 sobre a "memória contígua" dos vetores, o que o processador do seu servidor terá que fazer fisicamente com os outros 9.999.999 clientes quando o cliente 0 for removido da frente?

**Sua Resposta:**

Ocorrerá um **deslocamento em cascata de memória (Shift O(n))**.

Como a `list` do Python é implementada internamente como um **array contíguo** (os elementos ocupam posições sequenciais e adjacentes na RAM), remover o elemento do índice `0` cria um "buraco" no início do bloco de memória. Para manter a contiguidade, o processador é obrigado a **mover fisicamente todos os 9.999.999 elementos restantes uma posição para a esquerda** na memória.

Isso torna a operação `list.pop(0)` de complexidade **O(n)** — quanto maior a fila, mais lenta a operação. Em uma fila de 10 milhões de clientes, isso é catastrófico e pode travar o servidor.

A solução correta é usar `collections.deque` com `popleft()`, que opera em **O(1)** pois sua arquitetura interna (lista duplamente encadeada) apenas atualiza ponteiros, sem deslocar nada na memória.

```python
import tracemalloc

tracemalloc.start()

def fatorial_iterativo(n):
    fatorial = 1
    for i in range(1, n + 1):
        fatorial *= i
    return fatorial

print(fatorial_iterativo(100))

snapshot = tracemalloc.take_snapshot()

top_stats = snapshot.statistics('lineno')
for stat in top_stats[:5]:
    print(stat)

print(snapshot)

tracemalloc.stop()
```
